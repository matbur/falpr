```python
from glob import glob
import os

import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.set_cmap(plt.gray())

%matplotlib inline
```

```python
DIR = '../data/images_letters_normalized/A'

l = []
for i in glob(os.path.join(DIR, '*')):
    img = cv2.imread(i, 0)
    l.append(img)

len(l)
```

```python
plt.imshow(l[39])
l[0].shape
```

```python
mean = np.mean(l, 0)
plt.imshow(mean)
```

```python
def calculateDistance(i1, i2):
    return np.sum((i1-i2)**2)
```

```python
ll = sorted(l, key=lambda x: np.sum((x - mean)**2), reverse=1)

split_size = int(len(ll) / 10)
ll = ll[split_size:]
```

```python
n = 40
fig, axs = plt.subplots(1, n)
fig.set_size_inches(n, 10)
for i, v in enumerate(ll[:n]):
    axs[i].imshow(v)
fig.show()
```
